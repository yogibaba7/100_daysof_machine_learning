```python
# Import Libries
import pandas as pd 
import numpy as np 
```

```python
# import dataset
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
data.head()
```

```python
# Seperate input and output
X = data.drop(columns=['medv'])
y = data['medv']
```

```python
# KNN Model Training withOut seting any hyperparameter
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score,KFold

knn = KNeighborsRegressor()
kfold = KFold(n_splits=5,shuffle=True,random_state=1)
scores = cross_val_score(knn,X,y,cv=kfold,scoring='r2')

print('Mean Score:',scores.mean())
```

# GridSearchCv 

```python
# import gridsearchcv
from sklearn.model_selection import GridSearchCV
```

```python
# Decide Parameters Range

param_dict = {
    'n_neighbors' : [3,5,7,9,11,13,17,19],
    'weights' : ['uniform','distance'],
    'algorithm' : ['ball_tree', 'kd_tree', 'brute'],
    'p' : [1,2]
}

# Create model object 
knn = KNeighborsRegressor()
# Create KFold Object 
kfold = KFold(n_splits=5,shuffle=True,random_state=1)

# create gridsearch cv object 
gridsearchcv = GridSearchCV(knn,param_dict,cv=kfold,scoring='r2',refit=True,verbose=2)

# Fit The Data 
gridsearchcv.fit(X,y)

```

```python
# Best score -> Here, You can See By Adjusting Hyperparameter Model Performance Increased.
gridsearchcv.best_score_
```

```python
# Best Parameter
gridsearchcv.best_params_
```

```python
# Result 
result_df = pd.DataFrame(gridsearchcv.cv_results_)[['param_algorithm','param_n_neighbors','param_p','param_weights','mean_test_score','std_test_score']]
```

```python
result_df.sort_values(by ='mean_test_score',ascending=False)
```

# RandomizedSearchCV

```python
# import RandomSearchCV
from sklearn.model_selection import RandomizedSearchCV

# Decide Parameters Range

param_dict = {
    'n_neighbors' : [3,5,7,9,11,13,17,19],
    'weights' : ['uniform','distance'],
    'algorithm' : ['ball_tree', 'kd_tree', 'brute'],
    'p' : [1,2]
}

# Create model object 
knn = KNeighborsRegressor()
# Create KFold Object 
kfold = KFold(n_splits=5,shuffle=True,random_state=1)

# create gridsearch cv object 
randomizedsearchcv = RandomizedSearchCV(knn,param_dict,cv=kfold,scoring='r2',refit=True,verbose=2)

# Fit The Data 
randomizedsearchcv.fit(X,y)

```

```python
# Best Score
randomizedsearchcv.best_score_
```

```python
# Best Parameters
randomizedsearchcv.best_params_
```

```python
# Result 
pd.DataFrame(randomizedsearchcv.cv_results_)[['param_weights','param_p','param_n_neighbors','param_algorithm','mean_test_score']].sort_values(by='mean_test_score',ascending=False)
```
